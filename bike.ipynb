Extracting the dataset from the downloaded zip file and readong it into a pandas dataframe

```python
cols ={
    'timestamp':'time',
    'cnt':'count', 
    't1':'temp_real_C',
    't2':'temp_feels_like_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

# Renaming the columns to the specified column names in the dict
bikes_df.rename(cols, axis=1, inplace=True)
```

```python
bikes_df
```

```python
# Changing the humidity values to decimal ( a value between 0 and 1)
bikes_df.humidity_percent = bikes_df.humidity_percent / 100
bikes_df.humidity_percent
```

```python
# Season dict
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}

# Weather dict
weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}

# Changing the seasons column data type to string
bikes_df.season = bikes_df.season.astype('str')
# Mapping the values 0-3 to the actual written seasons
bikes_df.season = bikes_df.season.map(season_dict)

# Changing the weather column data type to string
bikes_df.weather = bikes_df.weather.astype('str')
# Mapping the values to the actual written weathers
bikes_df.weather = bikes_df.weather.map(weather_dict)
```
